```python
# Convertendo a coluna 'Data' para datetime e Ordenando o DataFrame pela coluna de datas
df1 = pd.read_excel('winfut_Renko12-04-23.xlsx');
df1['Data'] = pd.to_datetime(df1['Data'], dayfirst=True)  # dayfirst=True indica que o formato é dd/mm/aaaa
df1 = df1.sort_values(by='Data', ascending=True)

df2 = pd.read_excel('winfut_Renko14-06-23.xlsx')
df2['Data'] = pd.to_datetime(df2['Data'], dayfirst=True)
df2 = df2.sort_values(by='Data', ascending=True)

df3 = pd.read_excel('winfut_Renko16-08-23.xlsx')
df3['Data'] = pd.to_datetime(df3['Data'], dayfirst=True)
df3 = df3.sort_values(by='Data', ascending=True)

df4 = pd.read_excel('winfut_Renko18-10-23.xlsx')
df4['Data'] = pd.to_datetime(df4['Data'], dayfirst=True)
df4 = df4.sort_values(by='Data', ascending=True)
```

```python
df1['Origem'] = 'winfut_Renko12-04-23'
df2['Origem'] = 'winfut_Renko14-06-23'
df3['Origem'] = 'winfut_Renko16-08-23'
df4['Origem'] = 'winfut_Renko18-10-23'

winfutrenko = pd.concat([df1, df2, df3, df4], ignore_index=False)
```

```python
# Definindo a coluna 'Data' como índice
winfutrenko = winfutrenko.set_index('Data')
```

```python
winfutrenko.plot.line(y="Fechamento", use_index=True)
```

```python
winfutrenko["PredictTrade"] = winfutrenko["Fechamento"].shift(-1)
```

```python
winfutrenko["Alvo"] = (winfutrenko["PredictTrade"] > winfutrenko["Fechamento"]).astype(int)
```

```python
winfutrenko
```

```python
from sklearn.ensemble import RandomForestClassifier

# Criandr o modelo RandomForestClassifier com os parâmetros desejados
model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

# Dividindo os dados em conjuntos de treinamento e teste
train = winfutrenko.iloc[:-100]
test = winfutrenko.iloc[-100:]

# Especificando os preditores
predictors = ["Fechamento", "Volume Quantidade", "Abertura", "Máxima", "Mínima"]

# Ajustando o modelo aos dados de treinamento
model.fit(train[predictors], train["Alvo"])
```

```python
from sklearn.metrics import precision_score

# Realizando previsões com base nos dados de teste
preds = model.predict(test[predictors])

# Criando uma série pandas com as previsões e índice correspondente
preds = pd.Series(preds, index=test.index)

# Calculo de precisão comparando as previsões com os valores reais
precision_score(test["Alvo"], preds)
```

```python
# DataFrame combinado com as colunas "Alvo" e "Previsões"
combined = pd.concat([test["Alvo"], preds], axis=1)
combined.plot()
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Alvo"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Previsoes")
    combined = pd.concat([test["Alvo"], preds], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)

    return pd.concat(all_predictions)
```

```python
#Função backtest com o DataFrame winfutpredict, modelo e preditores
predictions = backtest(winfutrenko, model, predictors)
```

```python
horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = winfutrenko.rolling(horizon).mean()

    ratio_column = f"Close_Ratio_{horizon}"
    winfutrenko[ratio_column] = winfutrenko["Fechamento"] / rolling_averages["Fechamento"]

    trend_column = f"Trend_{horizon}"
    winfutrenko[trend_column] = winfutrenko.shift(1).rolling(horizon).sum()["Alvo"]

    new_predictors += [ratio_column, trend_column]
```

```python
winfutrenko = winfutrenko.dropna(subset=winfutrenko.columns[winfutrenko.columns != "PredictTrade"])
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Alvo"])
    preds = model.predict_proba(test[predictors])[:, 1]
    preds[preds >=.6] = 1
    preds[preds <.6] = 0
    preds = pd.Series(preds, index=test.index, name="Previsoes")
    combined = pd.concat([test["Alvo"], preds], axis=1)
    return combined
```

```python
# Exemplo de chamada da função backtest com o DataFrame winfutpredict, modelo e novos preditores
predictions = backtest(winfutrenko, model, new_predictors)
```
